```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
trig = np.linspace(-2*np.pi, 2*np.pi, 100)
```

# Plot

```python
plt.plot(trig,np.sin(trig))
```

Two graphs in same plot

```python
plt.plot(trig, np.sin(trig))
plt.plot(trig, np.cos(trig))
```

# Subplots

`Subplot of` $\sin(x), \cos(x), e^x, e^{-x}$

```python
plt.style.use("seaborn") # use different syle

# managing axes properties
fig = plt.figure()
fig.suptitle("Subplot Example")
plt.tight_layout()


plt.subplot(2,2,1)
plt.plot(trig, np.sin(trig), 'r-', label="$\sin(x)$")
plt.legend(loc="best")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(trig, np.cos(trig), 'b-', label="$\cos(x)$")
plt.legend(loc="best")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(trig, np.e**trig, 'g-', label="$e^x$")
plt.legend(loc="best")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(trig, np.e**-trig, 'black', label="$e^-x$")
plt.grid(True)
plt.legend(loc="best")
```
