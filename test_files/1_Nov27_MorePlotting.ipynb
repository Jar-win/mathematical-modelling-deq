```python
import matplotlib.pyplot as plt
import numpy as np
```

# 2. More Plotting

1. $y = sin(x)$

```python
x= np.linspace(0, 4*np.pi, 100)
fig= plt.figure(figsize=(10, 5))

one= plt.subplot(221)
plt.xlabel('x')     
plt.ylabel('sin(x)')
plt.plot(x, np.sin(x))
one.set_title("sine of x")
plt.grid(True)

two=plt.subplot(222)
plt.xlabel('x')     
plt.ylabel('cos(x)')
plt.plot(x, np.cos(x))
two.set_title("y= cosine of x")
plt.grid(True)

plt.subplot(223)
plt.xlabel('x')     
plt.ylabel('e^x')
plt.plot(x, np.e**x)
plt.grid(True)

plt.subplot(224)
plt.xlabel('x')     
plt.ylabel('e^(-x)')
plt.plot(x, np.e**(-x))
plt.grid(True)

plt.subplots_adjust(hspace=0.5, wspace= 0.4)
fig.suptitle("Some graphs")

plt.show()
```
