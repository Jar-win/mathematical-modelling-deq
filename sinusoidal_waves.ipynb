```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
```

```python
def plot_options(plot_list):
    for plot in plot_list:
        plt.plot(plot[0], plot[1], plot[2], label=plot[3])
```

```python
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plotlist = [[x, np.sin(x), 'r-', "$\sin(x)$"], 
            [x, 2*np.sin(x), 'g--', "$2 \sin(x)$"],
            [x, 4*np.sin(x), 'b', "$4 \sin(x)$"]]
plot_options(plotlist)

plt.legend(loc="best")
plt.grid(True)
plt.title("Change In Amplitude")

plt.subplot(2,2,2)
plt.plot(x, np.sin(2*x), 'r-', linewidth=0.7 ,label="$\sin(2x)$")
plt.plot(x, np.sin(2*x+1), 'g--',linewidth=0.7 , label="$\sin(2x+1)$")
plt.plot(x, np.sin(2*x+3), linewidth=0.7 ,label="$\sin(2x+3)$")
plt.legend(loc="best")
plt.title("Change in Angular Freq. and phase Difference")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x, np.sin(x), 'r-', linewidth=0.7 ,label="$\sin(x)$")
plt.plot(x, np.sin(x+2), 'g--', linewidth=0.7 ,label="$\sin(x+2)$")
plt.plot(x, np.sin(x+4),linewidth=0.7 , label="$\sin(x+4)$")
plt.legend(loc="best")
plt.title("Phase Difference")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x, np.sin(x), 'r-', linewidth=0.7 ,label="$\sin(x)$")
plt.plot(x, np.sin(2*x), 'g--', linewidth=0.7 ,label="$\sin(2x)$")
plt.plot(x, np.sin(3*x), 'b', linewidth=0.7 ,label="$\sin(3x)$")
plt.legend(loc="best")
plt.title("Change in Angualr Frequency")
plt.grid(True)
plt.savefig("Variable Frequency Amp Phase diff.png", dpi=500)
```
